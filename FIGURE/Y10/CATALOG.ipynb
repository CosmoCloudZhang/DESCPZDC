```python
import os
import h5py
import numpy
import corner
from matplotlib import pyplot
```

```python
BASE_FOLDER = '/global/cfs/cdirs/lsst/groups/MCP/CosmoCloud/DESCPZDC/'
CATALOG_FOLDER = os.path.join(BASE_FOLDER, 'CATALOG/')

TAG = 'Y10'
FIGURE_FOLDER = os.path.join(BASE_FOLDER, 'FIGURE/')
os.makedirs(os.path.join(FIGURE_FOLDER, '{}/'.format(TAG)), exist_ok=True)
```

```python
SIZE = 1000000
RANDOM = numpy.random.default_rng(seed=0)
os.environ['PATH'] = '/pscratch/sd/y/yhzhang/texlive/2025/bin/x86_64-linux:' + os.environ['PATH']
LABEL_LIST = [r'$z_\mathrm{spec}$', r'$m_i$', r'$u - g$', r'$g - r$', r'$r - i$', r'$i - z$', r'$z - y$']
RANGE_LIST = [(0.0, 3.0), (16.0, 26.0), (-1.5, +4.5), (-1.6, +3.2), (-1.2, +2.4), (-2.0, +2.0), (-1.0, +1.5)]

pyplot.rcParams['font.size'] = 25
pyplot.rcParams['text.usetex'] = True
pyplot.rcParams['pgf.texsystem'] = 'pdflatex'
pyplot.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'
```

```python
DIFFSKY = h5py.File(os.path.join(CATALOG_FOLDER, '{}/DIFFSKY.hdf5'.format(TAG)), 'r')
COSMODC2 = h5py.File(os.path.join(CATALOG_FOLDER, '{}/COSMODC2.hdf5'.format(TAG)), 'r')
POPCOSMOS = h5py.File(os.path.join(CATALOG_FOLDER, '{}/POPCOSMOS.hdf5'.format(TAG)), 'r')
OPENUNIVERSE = h5py.File(os.path.join(CATALOG_FOLDER, '{}/OPENUNIVERSE.hdf5'.format(TAG)), 'r')
```

```python
DIFFSKY_REDSHIFT = DIFFSKY['redshift'][...]
DIFFSKY_MAG_U_LSST = DIFFSKY['mag_u_lsst'][...]
DIFFSKY_MAG_G_LSST = DIFFSKY['mag_g_lsst'][...]
DIFFSKY_MAG_R_LSST = DIFFSKY['mag_r_lsst'][...]
DIFFSKY_MAG_I_LSST = DIFFSKY['mag_i_lsst'][...]
DIFFSKY_MAG_Z_LSST = DIFFSKY['mag_z_lsst'][...]
DIFFSKY_MAG_Y_LSST = DIFFSKY['mag_y_lsst'][...]

INDEX = RANDOM.choice(len(DIFFSKY_REDSHIFT), size=SIZE, replace=False)

DIFFSKY_REDSHIFT = DIFFSKY_REDSHIFT[INDEX]
DIFFSKY_MAG_U_LSST = DIFFSKY_MAG_U_LSST[INDEX]
DIFFSKY_MAG_G_LSST = DIFFSKY_MAG_G_LSST[INDEX]
DIFFSKY_MAG_R_LSST = DIFFSKY_MAG_R_LSST[INDEX]
DIFFSKY_MAG_I_LSST = DIFFSKY_MAG_I_LSST[INDEX]
DIFFSKY_MAG_Z_LSST = DIFFSKY_MAG_Z_LSST[INDEX]
DIFFSKY_MAG_Y_LSST = DIFFSKY_MAG_Y_LSST[INDEX]

DIFFSKY_DATA = numpy.stack([
    DIFFSKY_REDSHIFT, 
    DIFFSKY_MAG_I_LSST, 
    DIFFSKY_MAG_U_LSST - DIFFSKY_MAG_G_LSST, 
    DIFFSKY_MAG_G_LSST - DIFFSKY_MAG_R_LSST, 
    DIFFSKY_MAG_R_LSST - DIFFSKY_MAG_I_LSST, 
    DIFFSKY_MAG_I_LSST - DIFFSKY_MAG_Z_LSST, 
    DIFFSKY_MAG_Z_LSST - DIFFSKY_MAG_Y_LSST
], axis=1)
```

```python
POPCOSMOS_REDSHIFT = POPCOSMOS['redshift'][...]
POPCOSMOS_MAG_U_LSST = POPCOSMOS['mag_u_lsst'][...]
POPCOSMOS_MAG_G_LSST = POPCOSMOS['mag_g_lsst'][...]
POPCOSMOS_MAG_R_LSST = POPCOSMOS['mag_r_lsst'][...]
POPCOSMOS_MAG_I_LSST = POPCOSMOS['mag_i_lsst'][...]
POPCOSMOS_MAG_Z_LSST = POPCOSMOS['mag_z_lsst'][...]
POPCOSMOS_MAG_Y_LSST = POPCOSMOS['mag_y_lsst'][...]

INDEX = RANDOM.choice(len(POPCOSMOS_REDSHIFT), size=SIZE, replace=True)

POPCOSMOS_REDSHIFT = POPCOSMOS_REDSHIFT[INDEX]
POPCOSMOS_MAG_U_LSST = POPCOSMOS_MAG_U_LSST[INDEX]
POPCOSMOS_MAG_G_LSST = POPCOSMOS_MAG_G_LSST[INDEX]
POPCOSMOS_MAG_R_LSST = POPCOSMOS_MAG_R_LSST[INDEX]
POPCOSMOS_MAG_I_LSST = POPCOSMOS_MAG_I_LSST[INDEX]
POPCOSMOS_MAG_Z_LSST = POPCOSMOS_MAG_Z_LSST[INDEX]
POPCOSMOS_MAG_Y_LSST = POPCOSMOS_MAG_Y_LSST[INDEX]

POPCOSMOS_DATA = numpy.stack([
    POPCOSMOS_REDSHIFT, 
    POPCOSMOS_MAG_I_LSST, 
    POPCOSMOS_MAG_U_LSST - POPCOSMOS_MAG_G_LSST, 
    POPCOSMOS_MAG_G_LSST - POPCOSMOS_MAG_R_LSST, 
    POPCOSMOS_MAG_R_LSST - POPCOSMOS_MAG_I_LSST, 
    POPCOSMOS_MAG_I_LSST - POPCOSMOS_MAG_Z_LSST, 
    POPCOSMOS_MAG_Z_LSST - POPCOSMOS_MAG_Y_LSST
], axis=1)
```

```python
figure, plot = pyplot.subplots(nrows=len(LABEL_LIST), ncols=len(LABEL_LIST), figsize=(len(LABEL_LIST) * 3, len(LABEL_LIST) * 3))

figure = corner.corner(
    bins=25, 
    fig=figure,
    smooth=True,
    color='darkred',
    range=RANGE_LIST,
    data=DIFFSKY_DATA,
    labels=LABEL_LIST,
    hist_bin_factor=2, 
    show_titles=False, 
    plot_contours=True, 
    use_math_text=True, 
    hist_kwargs={
        'density': True,
        'histtype': 'step',
    },
    hist2d_kwargs={
        'plot_density': True,
        'plot_contours': True, 
        'fill_contours': False, 
        'plot_datapoints': False,
    },
)

figure = corner.corner(
    bins=25, 
    fig=figure,
    smooth=True,
    color='darkcyan',
    range=RANGE_LIST,
    labels=LABEL_LIST,
    hist_bin_factor=2, 
    show_titles=False,
    data=COSMODC2_DATA,
    plot_contours=True, 
    use_math_text=True, 
    hist_kwargs={
        'density': True,
        'histtype': 'step',
    },
    hist2d_kwargs={
        'plot_density': True,
        'plot_contours': True, 
        'fill_contours': False, 
        'plot_datapoints': False,
    },
)

for axes in figure.get_axes():
    axes.set_rasterized(True)

figure.savefig(os.path.join(FIGURE_FOLDER, '{}/COSMODC2.pdf'.format(TAG)), format='pdf', bbox_inches='tight')
pyplot.close(figure)
```
